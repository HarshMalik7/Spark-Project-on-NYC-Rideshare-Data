```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
january_data = pd.read_csv('january-data.csv')
```

```python
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
```

```python
plt.bar(january_data['day'], january_data['avg(request_to_pickup)'])
plt.xticks(january_data['day'])
plt.title("Histogram of Average Waiting Time during January")
plt.xlabel('Days of January')
plt.ylabel('Average Waiting Time')
plt.savefig('january-data.jpg')
```
