```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uber_data = pd.read_csv('Uber_data-part3.csv')
lyft_data = pd.read_csv('Lyft_data-part3.csv')
uber_data = uber_data.sort_values('business-date')
lyft_data = lyft_data.sort_values('business-date')
```

```python
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
```

```python
plt.bar(uber_data['business-date'], uber_data['sum(driver_total_pay)'])
xlocs, xlabs = plt.xticks()
xlocs=[i+1 for i in range(0,3130051)]
xlabs=[i/2 for i in range(0,3130051)]
for i, v in enumerate(uber_data['sum(driver_total_pay)']):
    plt.text(xlocs[i] - 1.4, v + 500000, str(v))
plt.title("Histogram of Driver's Earnings/Month for Uber")
plt.xlabel('Business-Month')
plt.ylabel('Total Amount')
plt.savefig('uber_data_part3.jpg')
```

```python
plt.bar(lyft_data['business-date'], lyft_data['sum(driver_total_pay)'])
xlocs, xlabs = plt.xticks()
xlocs=[i+1 for i in range(0,360408)]
xlabs=[i/2 for i in range(0,360408)]
for i, v in enumerate(lyft_data['sum(driver_total_pay)']):
    plt.text(xlocs[i] - 1.5, v, str(v))
plt.title("Histogram of Driver's Earnings/Month for Lyft")
plt.xlabel('Business-Month')
plt.ylabel('Count')
plt.savefig('lyft-data-part3.jpg')
```
