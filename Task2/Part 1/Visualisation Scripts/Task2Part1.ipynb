```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uber_data = pd.read_csv('Uber_data.csv')
lyft_data = pd.read_csv('Lyft_data.csv')
uber_data = uber_data.sort_values('business-date')
lyft_data = lyft_data.sort_values('business-date')
```

```python
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
```

```python
plt.bar(uber_data['business-date'], uber_data['count'])
xlocs, xlabs = plt.xticks()
xlocs=[i+1 for i in range(0,max(uber_data['count']))]
xlabs=[i/2 for i in range(0,max(uber_data['count']))]
for i, v in enumerate(uber_data['count']):
    plt.text(xlocs[i] - 1.4, v + 50000, str(v))
plt.title('Histogram of Business/Month for Uber')
plt.xlabel('Business-Month')
plt.ylabel('Count')
plt.savefig('uber_data.jpg')
```

```python
plt.bar(lyft_data['business-date'], lyft_data['count'])
xlocs, xlabs = plt.xticks()
xlocs=[i+1 for i in range(0,max(lyft_data['count']))]
xlabs=[i/2 for i in range(0,max(lyft_data['count']))]
for i, v in enumerate(lyft_data['count']):
    plt.text(xlocs[i] - 1.2, v, str(v))
plt.title('Histogram of Business/Month for Lyft')
plt.xlabel('Business-Month')
plt.ylabel('Count')
plt.savefig('lyft-data.jpg')
```
