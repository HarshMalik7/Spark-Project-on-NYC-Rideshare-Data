```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
uber_data = pd.read_csv('Uber-data_part2.csv')
lyft_data = pd.read_csv('Lyft-data_part2.csv')
uber_data = uber_data.sort_values('business-date')
lyft_data = lyft_data.sort_values('business-date')
```

```python
sns.set_theme(rc={'figure.figsize':(11.7,8.27)})
```

```python
plt.bar(uber_data['business-date'], uber_data['sum(rideshare_profit)'])
xlocs, xlabs = plt.xticks()
xlocs=[i+1 for i in range(0,1631336)]
xlabs=[i/2 for i in range(0,1631336)]
for i, v in enumerate(uber_data['sum(rideshare_profit)']):
    plt.text(xlocs[i] - 1.4, v + 50000, str(v))
plt.title('Histogram of Profit/Month for Uber')
plt.xlabel('Business-Month')
plt.ylabel('Total Amount')
plt.savefig('uber_data_part2.jpg')
```

```python
fig, ax = plt.subplots()
bars = ax.bar(lyft_data['business-date'], lyft_data['sum(rideshare_profit)'])
for bar in bars:
    yval = bar.get_height()
    if yval < 0:
        ax.text(bar.get_x() + bar.get_width() / 2, yval - 2, f'{yval}', ha='center', va='top')
    else:
        ax.text(bar.get_x() + bar.get_width() / 2, yval + 2, f'{yval}', ha='center', va='bottom')

plt.title('Histogram of Profit/Month for Lyft')
plt.xlabel('Business-Month')
plt.ylabel('Count')
plt.savefig('lyft-data_part2.jpg')
```
